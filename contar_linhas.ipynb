```python
import os
import pandas as pd

# Definir o caminho do diretório onde os arquivos estão localizados
diretorio = ''

# Inicializar uma lista para armazenar os dados
dados = []

# Loop para iterar sobre os arquivos de Meta_1_1.fq.filtered até Meta_59_1.fq.filtered
for i in range(1, 60):
    nome_arquivo = f'Meta_{i}_1.fq.filtered'
    caminho_arquivo = os.path.join(diretorio, nome_arquivo)
    
    if os.path.isfile(caminho_arquivo):
        # Contar as linhas do arquivo
        with open(caminho_arquivo, 'r') as file:
            num_linhas = sum(1 for line in file)
        
        # Adicionar os dados na lista
        dados.append([nome_arquivo, num_linhas])
        print(f'arquivo: {nome_arquivo} quant de linha { num_linhas}')
    else:
        # Caso o arquivo não exista, adicionar zero na contagem de linhas
        dados.append([nome_arquivo, 0])
        print(f'{nome_arquivo} 0')

# Criar um DataFrame do pandas com os dados
df = pd.DataFrame(dados, columns=['Arquivo', 'Quantidade de Linhas'])

# Exibir a tabela
print(df)

# Salvar a tabela em um arquivo CSV (opcional)
df.to_csv('contagem_de_linhas.csv', index=False)
```
